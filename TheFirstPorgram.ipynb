```python
from tkinter import *
from pyautocad import Autocad, APoint
a = Autocad()

#Tạo bảng
bang1 = Tk()
bang1.title('Nguyễn Văn Dũng')
bang1.geometry('500x400')
bang1.attributes("-topmost", True)

#Tạo label1
dong1 = Label(bang1, #dong1 là nhãn trong bang1 
              text = 'Tọa độ x điểm p1', 
              font = ('Times new roman',10))
dong1.place(x = 5, y = 5)

#Tạo entry1
onhap1 = Entry (bang1, 
               width= 5, 
               font= ('times new roman', 15))
onhap1.place(x=100, y=5)
onhap1.focus()

#Tạo label2
dong2 = Label(bang1, #dong1 là nhãn trong bang1 
              text = 'Tọa độ y điểm p1', 
              font = ('Times new roman',10))
dong2.place(x = 200, y = 5)

#Tạo entry2
onhap2 = Entry (bang1, 
               width= 5, 
               font= ('times new roman', 15))
onhap2.place(x=300, y=5)


#Tạo label3
dong3 = Label(bang1, #dong1 là nhãn trong bang1 
              text = 'Tọa độ x điểm p2', 
              font = ('Times new roman',10))
dong3.place(x = 5, y = 50)

#Tạo entry3
onhap3 = Entry (bang1, 
               width= 5, 
               font= ('times new roman', 15))
onhap3.place(x=100, y=50)
onhap3.focus()

#Tạo label4
dong4 = Label(bang1, #dong1 là nhãn trong bang1 
              text = 'Tọa độ y điểm p2', 
              font = ('Times new roman',10))
dong4.place(x = 200, y = 50)

#Tạo entry4
onhap4 = Entry (bang1, 
               width= 5, 
               font= ('times new roman', 15))
onhap4.place(x=300, y=50)

#Tạo label5
dong5 = Label(bang1, #dong1 là nhãn trong bang1 
              text = 'Màu của đoạn thẳng', 
              font = ('Times new roman',10))
dong5.place(x = 5, y = 100)

#Tạo entry5
onhap5 = Entry (bang1, 
               width= 5, 
               font= ('times new roman', 15))
onhap5.place(x=150, y=100)

#Tạo hàm anvao
def anvao(): 
    p1 = APoint(float(onhap1.get()), float(onhap2.get()))
    p2 = APoint(float(onhap3.get()), float(onhap4.get()))
    line1 = a.model.AddLine(p1,p2)
    line1.color = int(onhap5.get())
    a.app.ZoomExtents()
    #Tạo label đã vẽ xong
    dong6 = Label(bang1,
              text = 'Đã vẽ xong',
              fg= 'red',
                bg='yellow',
              font = ('Times new roman',10))
    dong6.place(x = 170, y = 270)
    a.prompt('Đã vẽ xong')
#Tạo nút bấm
but1 = Button(bang1, 
             text = 'Vẽ ngay và luôn',  
             font = ('times new roman', 13), 
             width = 15, height = 2,
             command= anvao)
but1.place( x = 130, y = 200)
but1.mainloop()


bang1.mainloop()
```
