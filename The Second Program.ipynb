```python
from pyautocad import Autocad, APoint, aDouble
from tkinter import *
from PIL import ImageTk, Image
a=Autocad()

#Tạo tkinter

giaodien = Tk()
giaodien.title('Phần mềm vẽ T')
giaodien.geometry('400x600')

#import hình ảnh:

ha_import = (Image.open(r'C:\Users\DELL\Desktop\Python CAD\T.PNG'))
ha_size = ha_import.resize((250,250), Image.LANCZOS)
img = ImageTk.PhotoImage(ha_size)

#Import hình ảnh vào button
hinhanh = Button(giaodien, 
                 text ='', 
                 font=('times new roman', 13), 
                 image = img)
hinhanh.place(x=75, y=250)
 
#Label 'Nhập vào giá trị B'
dong1 = Label(giaodien, 
              text='Nhập vào giá trị B', 
              font=('times new roman', 12))
dong1.place(x=20,y=20)

#Label 'Nhập vào giá trị H'
dong2 = Label(giaodien, 
              text='Nhập vào giá trị H', 
              font=('times new roman', 12))
dong2.place(x=20,y=50)

#Label 'Nhập vào giá trị b'
dong3 = Label(giaodien, 
              text='Nhập vào giá trị b', 
              font=('times new roman', 12))
dong3.place(x=20,y=80)

#Label 'Nhập vào giá trị c'
dong4 = Label(giaodien, 
              text='Nhập vào giá trị H', 
              font=('times new roman', 12))
dong4.place(x=20,y=110)

#tạo ô nhập B
onhap1 = Entry (giaodien, 
               width= 10, 
               font= ('times new roman', 12))
onhap1.insert(END, '200')
onhap1.place(x=150,y=20)

#tạo ô nhập H
onhap2 = Entry (giaodien, 
               width= 10, 
               font= ('times new roman', 12))
onhap2.insert(END, '300')
onhap2.place(x=150,y=50)

#tạo ô nhập b
onhap3 = Entry (giaodien, 
               width= 10, 
               font= ('times new roman', 12))
onhap3.insert(END, '50')
onhap3.place(x=150,y=80)

#tạo ô nhập c
onhap4 = Entry (giaodien, 
               width= 10, 
               font= ('times new roman', 12))
onhap4.insert(END, '50')
onhap4.place(x=150,y=110)

#def vengay
def vengay():
    B=float(onhap1.get())
    H=float(onhap2.get())
    b=float(onhap3.get())
    c=float(onhap4.get())



    p1 = aDouble(0,0,0, 
                 b,0,0, 
                 b, H - c,0, 
                 b+(B-b)/2, H - c ,0, 
                 b+(B-c)/2, H,0,
                -(B-b)/2, H,0,
                -(B-b)/2, H - c ,0,
                0, H - c ,0,
                0,0,0)
    b1 = a.model.AddPolyline(p1)
    dim1= a.model.AddDimAligned(APoint(0),APoint(b,0),APoint(0,-10))


    s = b1.area
    #Label Diện tích hình đã vẽ là
    dientich = Label(giaodien, 
              text='Diện tích hình đã vẽ là: ' + str(s), 
              font=('times new roman', 12))
    dientich.place(x=20,y=530)

    def xoangay():
        b1.erase()
        dim1.erase()
        dientich.place(x=999,y=999)

    but2 = Button(giaodien, text = 'Xóa ngay', width = 10, height = 3,
              command= xoangay)
    but2.place( x = 250, y = 90)

but1 = Button(giaodien, text = 'Vẽ ngay', width = 10, height = 3,
              command= vengay)
but1.place( x = 250, y = 40)



mainloop()
```
